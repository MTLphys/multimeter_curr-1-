```python
import numpy as np 
import matplotlib.pyplot as plt
N= 10000
x= np.linspace(0,20*np.pi,10000)
y= np.sin(x*19*2*np.pi)#+np.random.rand(N)*1
window = np.kaiser(len(x),14)


#noise =0 
yn = y+0*2*(np.random.rand(N)-.5)
y1= np.round(4095*(1+yn)/2)*2/4095-1
cn1 = np.fft.fft(y1)
cn1w = np.fft.fft(y1*window)

#noise =.01 
yn = y+0.01*2*(np.random.rand(N)-.5)
y2= np.round(4095*(1+yn)/2)*2/4095-1
cn2 = np.fft.fft(y2)
cn2w = np.fft.fft(y2*window)

#noise =.05 
yn = y+0.05*2*(np.random.rand(N)-.5)
y4= np.round(4095*(1+yn)/2)*2/4095-1
cn4 = np.fft.fft(y4)
cn4w = np.fft.fft(y4*window)

#noise =.1 
yn = y+0.1*2*(np.random.rand(N)-.5)
y8= np.round(4095*(1+yn)/2)*2/4095-1
cn8 = np.fft.fft(y8)
cn8w = np.fft.fft(y8*window)

#noise =.5
yn = y+0.5*2*(np.random.rand(N)-.5)
y12= np.round(4095*(1+yn)/2)*2/4095-1
cn12 = np.fft.fft(y12)
cn12w = np.fft.fft(y12*window)

#noise =1
yn = y+2*2*(np.random.rand(N)-.5)
yn=np.round(4095*(1+yn)/2)*2/4095-1
cn = np.fft.fft(yn)
cnw = np.fft.fft(yn*window)
fn = np.fft.fftfreq(len(x),x[0]-x[1])

fig,ax= plt.subplots(3)

ax[0].plot(x,y1)
ax[0].plot(x,y2)
ax[0].plot(x,y4)
ax[0].plot(x,y8)
ax[0].plot(x,y12)
ax[0].plot(x,yn)
ax[0].set_xlim(0,.5)

ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn1)),label = 'noise=0')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn2)),label = 'noise=.01')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn4)),label = 'noise=.05')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn8)),label = 'noise=.1')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn12)),label = 'noise=.5')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn)),label = 'noise=1.0')

ax[1].set_title('unwindowed')

ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn1w)),label = 'noise=0')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn2w)),label = 'noise=.01')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn4w)),label = 'noise=.05')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn8w)),label = 'noise=.1')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn12w)),label = 'noise=.5')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cnw)),label = 'noise=1.0')
ax[2].set_title('windowed')
ax[1].legend()
#ax[2].legend()
ax[1].set_xlim(14,23)
ax[1].set_yscale('log')
ax[1].set_ylim(1e-4,1e5)
ax[2].set_xlim(14,23)
ax[2].set_yscale('log')
ax[2].set_ylim(1e-4,1e5)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.constants as c 

def freqwidth(wl,Dwl):
    wlm=(wl-Dwl)*1e-9
    wlp=(wl+Dwl)*1e-9
    return (c.c/wlm-c.c/wlp)/1e6 
centroid = (c.c/820e-9)/1e6
Domega = 8000#frequency spacing (Mhz)
domega = freqwidth(820,10) #pulse spectral width(Thz)
fullwindow = domega*10 
N=int(np.round(fullwindow/Domega))
fn =np.linspace(centroid-fullwindow/2,centroid+fullwindow/2,N+1)
#fn = np.linspace()
#Ans= 
print(domega)
AnR= np.exp(-.5*((fn-centroid)/(domega))**2)
#plt.plot(fn,AnR)

w = 10/Domega
N = int(np.round(w/centroid))
t = np.linspace(0,w,N)

out = np.zeros(N,dtype=np.complex128)
for i,fi in enumerate(fn):
    out+=AnR[i]*np.exp(1.0j*2*np.pi*t*fi)

fig,axs= plt.subplots(2)
ax[0].plot(fn,AnR)

#ax[1].plot(t,out)
```
