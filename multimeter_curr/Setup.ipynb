```python
import numpy as np 
import matplotlib.pyplot as plt
N= 10000
x= np.linspace(0,8*np.pi,10000)
y= np.sin(x*19*2*np.pi)#+np.random.rand(N)*1
window = np.kaiser(len(x),14)
#N= 64
cn = np.fft.fft(y)
cnw = np.fft.fft(y*window)
fn = np.fft.fftfreq(len(x),x[0]-x[1])

#N =1 
y1= np.round((1+y)/2)*2-1
cn1 = np.fft.fft(y1)
cn1w = np.fft.fft(y1*window)

#N =2
y2= np.round(3*(1+y)/2)*2/3-1
cn2 = np.fft.fft(y2)
cn2w = np.fft.fft(y2*window)

#N =4
y4= np.round(15*(1+y)/2)*2/15-1
cn4 = np.fft.fft(y4)
cn4w = np.fft.fft(y4*window)

#N = 8
y8= np.round(255*(1+y)/2)*2/255-1
cn8 = np.fft.fft(y8)
cn8w = np.fft.fft(y8*window)

#N= 12 
y12= np.round(4095*(1+y)/2)*2/4095-1
cn12 = np.fft.fft(y12)
cn12w = np.fft.fft(y12*window)

fig,ax= plt.subplots(3)
ax[0].plot(x,y)
ax[0].plot(x,y1)
ax[0].plot(x,y2)
ax[0].plot(x,y4)
ax[0].plot(x,y8)
ax[0].plot(x,y12)

ax[0].set_xlim(0,.5)

ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn1)),label = 'N=1')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn2)),label = 'N=2')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn4)),label = 'N=4')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn8)),label = 'N=8')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn12)),label = 'N=12')
ax[1].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn)),label = 'N=64')

ax[1].set_title('unwindowed')

ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn1w)),label = 'N=1')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn2w)),label = 'N=2')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn4w)),label = 'N=4')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn8w)),label = 'N=8')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cn12w)),label = 'N=12')
ax[2].plot(np.fft.fftshift(fn),np.fft.fftshift(np.abs(cnw)),label = 'N=64')
ax[2].set_title('windowed')
ax[1].legend()
#ax[2].legend()
ax[1].set_xlim(14,23)
ax[1].set_yscale('log')
ax[1].set_ylim(1e-1,1e5)
ax[2].set_xlim(14,23)
ax[2].set_yscale('log')
ax[2].set_ylim(1e-4,1e5)
```
